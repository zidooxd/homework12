#Постановка задач
<a name = 'Постановка_задач'>

##Задания:

1. Установить GIT по ссылке https://gitforwindows.org/
2. Создать локальный репозиторий.
3. Создать файл отчета по выполнению практической работы и поместить его в репозиторий
4. Зафиксируйте в отчете все проделанные шаги и создайте версию отчета в GIT
5. Создайте ветку отчета в GIT, добавьте описания шагов в отчет и создайте версию отчета в новой ветке.
6. Слейте последнее изменение отчета в основную ветку.
7. Регистрируемся на github (или используем уже существующий аккаунт), создаем проект «homework IIT», заливаем содержимое из локального репозитория в проект на github.
8. Разместить на github ранее созданный статический сайт. (Доп. задание – Разметить на github ранее созданный динамический сайт.)
9. Завершить отчет, добавив описание всех пунктов.




##Цель работы

Локальный репозиторий

##Расчет входных данных

1. GIT

# Ход работы
<a name = 'Ход_работы'>


Устанавливаем GIT SCM.

image.png

Указываем пусть с помощью команды $ cd /home/user/project, в моем случае я использую путь C:/Users/mirza/my_project/. Создаем репозиторий командой git init.

image.png

Смотрим все находящиеся файлы с помощью команды find . -name *.(расширение файла). Создаем файл homework12 и помещаем его в репозиторий командой git add homework12.ipynb.

image.png

Фиксируем в отчете все проделанные шаги и создаем версию отчета в GIT.
Обновилась ли версия отчета мы можем проверить коммандой git status.

image.png

Убеждаемся что версия обновилась, делаем коммит в локальном репозитории (git commit –m) и переносим измененный файл командой git push.

image.png

Регистрируемся на github, создаем проект «homework12» и переносим весь наш репозиторий на GitHub. Для этого используем
git remote add origin https://github.com/zidooxd/homework12.git,
git branch -M main,
git push -u origin main.
